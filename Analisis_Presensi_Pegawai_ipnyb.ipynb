```
import numpy as np
import IPython.display as display
from matplotlib import pyplot as plt
import io
import base64

ys = 200 + np.random.randn(100)
x = [x for x in range(len(ys))]

fig = plt.figure(figsize=(4, 3), facecolor='w')
plt.plot(x, ys, '-')
plt.fill_between(x, ys, 195, where=(ys > 195), facecolor='g', alpha=0.6)
plt.title("Sample Visualization", fontsize=10)

data = io.BytesIO()
plt.savefig(data)
image = F"data:image/png;base64,{base64.b64encode(data.getvalue()).decode()}"
alt = "Sample Visualization"
display.display(display.Markdown(F"""![{alt}]({image})"""))
plt.close(fig)
```

```
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# 1. Load Data (Gunakan engine openpyxl untuk .xlsx)
file_name = 'data_pegawai.xlsx'
df = pd.read_excel(file_name)

# 2. Konversi kolom tanggal ke format datetime

df['Tgl Presensi'] = pd.to_datetime(df['Tgl Presensi'])

# 3. Ekstrak informasi waktu untuk layout kalender
df['Bulan'] = df['Tgl Presensi'].dt.month_name()
df['Minggu_ke'] = df['Tgl Presensi'].dt.strftime('%U').astype(int) # Nomor minggu dalam setahun
df['Hari'] = df['Tgl Presensi'].dt.day_name()
df['Tanggal'] = df['Tgl Presensi'].dt.day

# 4. Hitung jumlah kehadiran per hari
# Kita hitung berapa banyak baris (pegawai) yang hadir di tanggal tersebut
daily_data = df.groupby(['Bulan', 'Minggu_ke', 'Hari', 'Tanggal']).size().reset_index(name='Jumlah_Hadir')

# 5. Buat Pivot Table untuk layout Kalender
# Kita pilih satu bulan saja, misal 'March' (sesuaikan dengan data Anda)
bulan_pilihan = daily_data['Bulan'].unique()[0]
df_bulan = daily_data[daily_data['Bulan'] == bulan_pilihan]

days_order = ['Sunday', 'Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday']
pivot_cal = df_bulan.pivot(index='Minggu_ke', columns='Hari', values='Jumlah_Hadir')
pivot_cal = pivot_cal.reindex(columns=days_order)

# 6. Visualisasi dengan Heatmap agar menjadi Calendar View
plt.figure(figsize=(12, 8))
sns.heatmap(pivot_cal,
            annot=True, # Munculkan angka jumlah hadir
            fmt="g",
            cmap="RdYlGn", # Merah (rendah) ke Hijau (tinggi)
            linewidths=2,
            linecolor='white',
            cbar_kws={'label': 'Total Pegawai Hadir'})

plt.title(f'Calendar View Kehadiran Pegawai - {bulan_pilihan} 2025', fontsize=16, pad=20)
plt.xlabel('Hari', fontsize=12)
plt.ylabel('Minggu Ke-', fontsize=12)
plt.show()
```

```
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import math

# 1. Load Data
file_name = 'data_pegawai.xlsx'
df = pd.read_excel(file_name)

# 2. Konversi tanggal dan ekstrak fitur waktu
df['Tgl Presensi'] = pd.to_datetime(df['Tgl Presensi'])
df['Bulan'] = df['Tgl Presensi'].dt.month_name()
df['Bulan_Angka'] = df['Tgl Presensi'].dt.month
df['Tahun'] = df['Tgl Presensi'].dt.year
df['Minggu_ke'] = df['Tgl Presensi'].dt.strftime('%U').astype(int)
df['Hari'] = df['Tgl Presensi'].dt.day_name()

# 3. Hitung jumlah kehadiran per hari
daily_data = df.groupby(['Tahun', 'Bulan_Angka', 'Bulan', 'Minggu_ke', 'Hari']).size().reset_index(name='Jumlah_Hadir')

# Urutan hari agar standar kalender (Minggu ke Sabtu)
days_order = ['Sunday', 'Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday']

# 4. Pengaturan Subplots (Otomatis menyesuaikan jumlah bulan)
list_bulan = daily_data.sort_values(['Tahun', 'Bulan_Angka'])['Bulan'].unique()
n_bulan = len(list_bulan)
cols = 2 # Jumlah kolom tampilan (bisa diubah ke 3 atau 4)
rows = math.ceil(n_bulan / cols)

fig, axes = plt.subplots(rows, cols, figsize=(16, 5 * rows))
axes = axes.flatten() # Meratakan array axes agar mudah di-looping

# 5. Looping untuk membuat heatmap tiap bulan
for i, bulan in enumerate(list_bulan):
    data_per_bulan = daily_data[daily_data['Bulan'] == bulan]

    # Pivot data untuk bulan tersebut
    pivot_month = data_per_bulan.pivot(index='Minggu_ke', columns='Hari', values='Jumlah_Hadir')
    pivot_month = pivot_month.reindex(columns=days_order)

    # Gambar heatmap pada axis yang ditentukan
    sns.heatmap(pivot_month,
                annot=True,
                fmt="g",
                cmap="YlGnBu",
                ax=axes[i],
                cbar=False, # Matikan colorbar individual agar tidak penuh
                linewidths=1,
                linecolor='white')

    axes[i].set_title(f'Bulan: {bulan}', fontsize=14, fontweight='bold')
    axes[i].set_xlabel('')
    axes[i].set_ylabel('Minggu ke-')

# 6. Hapus subplot kosong jika jumlah bulan ganjil
for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])

plt.suptitle('Analisis Pola Presensi Pegawai Per Bulan (2025)', fontsize=20, y=1.02)
plt.tight_layout()
plt.show()
```

```
# import alat untuk visualisasi
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```
# masukkan file excel
df = pd.read_excel('data_pegawai.xlsx')
```

```
#Preprocessing tanggal agar teks biasa yang berupa data mentah menjadi objek waktu

df['Tgl Presensi'] = pd.to_datetime(df['Tgl Presensi']) # mengubah teks tipe data dari string jadi date time
df['Bulan'] = df['Tgl Presensi'].dt.month_name() # mengambil nama bulan dri tanggal itu
df['Hari'] = df['Tgl Presensi'].dt.day_name() # mengambil nama hari dari setiap baris data
df['Minggu_ke'] = df['Tgl Presensi'].dt.strftime('%U').astype(int) # menghitung urutan minggu dalam satu tahun ( 0-52)
df['Tanggal'] = df['Tgl Presensi'].dt.day # mengambil angka tanggal (1-31)
```

```
# Buatkan khusus untuk bulan januari terlebih dahulu
df_jan = df[df['Bulan'] == 'January'].copy()
```

```
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Load Data
df = pd.read_excel('data_pegawai.xlsx')

# 2. Persiapan Data
# Menghitung total kehadiran per Unit Kerja
# Pastikan nama kolom 'Unit' sesuai dengan file Anda (misal: 'Unit Kerja' atau 'Departemen')
nama_kolom_unit = 'Unit Kerja' # Diperbaiki dari 'Unit'
unit_comparison = df[nama_kolom_unit].value_counts().sort_values(ascending=True)

# 3. Visualisasi Horizontal Bar Chart
plt.figure(figsize=(10, 6))
# Menggunakan warna biru profesional yang konsisten
colors = '#2c3e50'

# Membuat grafik batang horizontal
bars = plt.barh(unit_comparison.index, unit_comparison.values, color=colors)

# Menambahkan label angka di ujung batang agar tidak misleading
for bar in bars:
    plt.text(bar.get_width() + 1, bar.get_y() + bar.get_height()/2,
             f'{int(bar.get_width())}', va='center', fontsize=10)

# 4. Ketentuan Desain (Judul & Label)
plt.title('Perbandingan Total Kehadiran per Unit Kerja (2025)', fontsize=16, pad=20, fontweight='bold')
plt.xlabel('Total Frekuensi Kehadiran (Hari-Orang)', fontsize=12)
plt.ylabel('Unit Kerja', fontsize=12)

# Menghilangkan garis grid yang tidak perlu (mengurangi clutter)
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)
plt.grid(axis='x', linestyle='--', alpha=0.5)

plt.tight_layout()
plt.show()
```

```
import pandas as pd
import matplotlib.pyplot as plt

# 1. Load Data
df = pd.read_excel('data_pegawai.xlsx')

# 2. Persiapan Data (Pastikan kolom 'Unit' benar)
# Menghitung total kehadiran per Unit Kerja
unit_comparison = df['Unit'].value_counts().sort_values(ascending=True)

# 3. Visualisasi dengan Ukuran yang Lebih Besar (Penting!)
# Menambah figsize (lebar, tinggi) agar baris punya ruang bernapas
plt.figure(figsize=(12, 10))

# Membuat grafik batang horizontal
# Menggunakan warna biru gelap yang konsisten dan elegan
bars = plt.barh(unit_comparison.index, unit_comparison.values, color='#2c3e50', height=0.7)

# 4. Menambahkan Label Angka di Ujung Batang
# Supaya tidak misleading, pimpinan bisa baca angka pastinya
for bar in bars:
    plt.text(bar.get_width() + 1,
             bar.get_y() + bar.get_height()/2,
             f' {int(bar.get_width())}',
             va='center', fontsize=11, fontweight='bold')

# 5. Desain Informatif & Anti-Clutter
plt.title('Perbandingan Total Kehadiran per Unit Kerja\nKampus XYZ - Tahun 2025',
          fontsize=16, pad=30, fontweight='bold')
plt.xlabel('Total Frekuensi Kehadiran (Hari-Orang)', fontsize=12, labelpad=15)
plt.ylabel('Unit Kerja', fontsize=12, labelpad=15)

# Menghilangkan garis tepi kanan dan atas agar bersih
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)

# Memberikan grid halus hanya pada sumbu X
plt.grid(axis='x', linestyle='--', alpha=0.3)

# Mengatur margin agar teks label tidak terpotong
plt.tight_layout()

plt.show()
```
